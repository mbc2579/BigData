### 데이터 타입 변경
분양가격이 object(문자) 타입으로 되어 있으므로 문자열 타입을 계산할 수 없기 때문에 수치 데이터로 변경 결측치가 섞여 있을 때 변환이 제대로 안됨으로 pd.to_numeric 을 통해 데이터의 타입을 변경

```python
type(np.nan)
```

```python
df_last["분양가격"] = pd.to_numeric(df_last["분양가격(㎡)"], errors='coerce')
df_last["분양가격"].head(1)
```

### 평당분양가격 구하기
공공데이터포털에 올라와 있는 2013년부터의 데이터는 평당분양가격 기준으로 되어 있음
분양가격을 평당기준으로 보기위해 3.3을 곱해서 "평당분양가격" 컬럼을 만들어 추가

```python
df_last["평당분양가격"] = df_last["분양가격"] * 3.3
df_last.head(1)
```

### 규모구분을 전용면적 컬럼으로 변경
str 의 replace 기능을 사용해서 예를들면 "전용면적 60㎡초과 85㎡이하"라면 "60㎡~85㎡" 로 변경



```python
df_last["규모구분"].unique()
```

```python
df_last["전용면적"] = df_last["규모구분"].str.replace("전용면적", "")
df_last["전용면적"] = df_last["전용면적"].str.replace("초과", "~")
df_last["전용면적"] = df_last["전용면적"].str.replace("이하", "")
df_last["전용면적"] = df_last["전용면적"].str.replace(" ", "").str.strip()
df_last["전용면적"]
```

### melt로 Tidy data 만들기
pandas의 melt를 사용하면 데이터의 형태를 변경할 수 있음

같은 형태로 만들어주어야 데이터를 합칠 수 있음
데이터를 병합하기 위해 melt를 사용



```python
df_first.head(1)
```

### 연도와 월을 분리하기
* pandas 의 string-handling 사용

```python
date = "2013년12월"
date
```

```python
# split 을 통해 "년"을 기준으로 텍스트를 분리
date.split("년")
```

```python
# 리스트의 인덱싱을 사용해서 연도만 가져옴
date.split("년")[0]
```

```python
# 리스트의 인덱싱과 replace를 사용해서 월을 제거

date.split("년")[-1].replace("월", "")
```

```python
# parse_year라는 함수를 만듬
# 연도만 반환하도록 하며, 반환하는 데이터는 int 타입이 되도록 함

def parse_year(date):
    year = date.split("년")[0]
    year = int(year)
    return year

y = parse_year(date)
print(type(y))
y
```

```python
# 제대로 분리가 되었는지 parse_year 함수를 확인
parse_year(date)
```

```python
# parse_month 라는 함수를 만듬
# 월만 반환하도록 하며, 반환하는 데이터는 int 타입이 되도록 함
def parse_month(date):
    month = date.split("년")[-1].replace("월", "")
    month = int(month)
    return month
```

```python
# 제대로 분리가 되었는지 parse_month 함수를 확인
parse_month(date)
```

```python
# df_first_melt 변수에 담긴 데이터프레임에서 
# apply를 활용해 연도만 추출해서 새로운 컬럼에 담음

df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt.head(1)
```

```python
# df_first_melt 변수에 담긴 데이터프레임에서 
# apply를 활용해 월만 추출해서 새로운 컬럼에 담음

df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)
df_first_melt.head(1)
```

```python
# df_last와 병합을 하기 위해서는 컬럼의 이름이 같아야 함
# sample을 활용해서 데이터를 미리보기
df_last.sample()
```
